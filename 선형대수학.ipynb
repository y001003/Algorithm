# 선형대수

## 기본용어

- 스칼라(Scalar): 숫자 하나로 구성된 데이터
- 벡터(Vector): 여러 숫자로 구성된 데이터 레코드

    방향이 있고 순서가 있고 차원의 공간에 존재하는 점  
    세로줄 열(column)은 하나, 가로줄 행(row)은 여러개인 형태로 위에서 아래로 써서 표기  
    벡터 하나를 이루는 데이터 개수가 n개이면 이 벡터를 'n-차원 벡터'라고 한다.

$$ x = \begin{bmatrix}1\\3\\ \end{bmatrix}$$
- 행렬(Matrix): 벡터가 여럿인 데이터 집합, 2차원 배열

    인공지능에서 행을 특성 행렬(feature matrix)라고 한다.
$$\begin{bmatrix}1&2\\3&4\\ \end{bmatrix}$$

- 텐서(Tensor): 크기가 같은 행렬이 여러개, 3차원 이상의 배열

## 기하학의 선형대수학

### 기하학에서의 벡터

'크기(magnitude)와 방향(direction)을 가진 물리량'  
'시작점에서 끝점을 연결한 화살표'

image.png

#### 벡터의 특징


1. 방향성을 갖는다.$\vec{a}$
    - 시작점(A)에서 끝점(B)으로 방향성 
    - 선분의 길이(r) = 벡터 크기
2. 크기와 방향이 중요하지 좌표상의 어디에 표현하는지 중요하지 않다.
3. 같은 벡터라고 해서 좌표상의 위치까지 동일하지는 않다.
4. 크기가 같더라도 방향이 다르면 다른 벡터이다.

#### 단위 벡터



단위 벡터 $\hat{v}$ : 크기가 1이 되도록 늘리거나 줄인 벡터

기하학적으로 같은 벡터라도 대수학적으로는 의미가 다를 수 있다. ➡ 좌표가 다른 기하학적으로 같은 벡터

☘ $ |\vec{v}|$ = 벡터의 크기 = 벡터의 길이 = 벡터의 norm(노름)
$$
단위 벡터 \hat{v} = \frac{\vec{v}}{|\vec{v}|}= \begin{bmatrix} \frac {x}{\sqrt {x^2 + y^2} } \ \frac {y}{\sqrt {x^2 + y^2} }\\ \end{bmatrix} 
$$

#### 위치 벡터

위치 벡터 : 원점이 시작점에 있지 않은 벡터를, 원점이 시작점인 벡터들로 표현
1.  원점 벡터를 시작점으로 한 벡터를 표현합니다.

2. 벡터의 뺄셈 연산을 합니다.

    ※ 뺄셈의 방향은 끝점에서 시작점 방향입니다.

3.  위치 벡터 성분을 도출합니다.

#### 영 벡터
모든 성분이 0으로 구성된 벡터
